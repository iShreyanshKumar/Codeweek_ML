```python
# Impute categories yes/no with mode
for col in ['famsize','higher']:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Impute 1-5 rating values with median
for col in ['Fedu', 'traveltime', 'freetime', 'absences', 'Feature_1', 'Feature_2', 'Feature_3']:
    df[col].fillna(df[col].median(), inplace=True)

# Impute grade with mean
df['G2'].fillna(df['G2'].mean(), inplace=True)

numeric_df = df.select_dtypes(include=np.number)
```

#Level 4
 Objective
 - Apply classification techniques to model relationship likelihood and assess their
performance. Think critically about what the models reveal, and what they donâ€™t

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
from scipy.stats import randint
from sklearn.model_selection import RandomizedSearchCV
from sklearn import model_selection
```

```python
features2 = numeric_df.drop(columns=['romantic']).columns
X = numeric_df[features2]
y = numeric_df['romantic']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

####Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sta = StandardScaler()

# Use .iloc for integer-location based indexing
X_train.iloc[:, :] = sta.fit_transform(X_train.iloc[:, :])
X_test.iloc[:, :] = sta.transform(X_test.iloc[:, :])
```

##Now running the classification model.

##1) Logistic Regression

Implementation- [https://www.datacamp.com/tutorial/understanding-logistic-regression-python](https://www.datacamp.com/tutorial/understanding-logistic-regression-python)

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(random_state=16)

# fit the model with data
logreg.fit(X_train, y_train)

y_pred = logreg.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

##2) Random Forest (with hyperparameter tuning)

Implementation (given in resources)- [https://www.analyticsvidhya.com/blog/2021/06/5-hyperparameter-optimization-techniques-you-must-know-for-data-science-hackathons/](https://www.analyticsvidhya.com/blog/2021/06/5-hyperparameter-optimization-techniques-you-must-know-for-data-science-hackathons/)

```python
#Importing XGBoost
from xgboost import XGBClassifier

#Defining XGB classification model
clf = XGBClassifier()
```

#### GridsearchCV

```python
#Importing packages from sklearn

from sklearn import preprocessing

from sklearn import model_selection

from sklearn import metrics

#defining a set of values as a dictionary for hyperparameters

param_grid = {

    "n_estimators":[100,200,300,400],

    "max_depth":[1,3,5,7],

    "reg_lambda":[.01,.1,.5]

}

#declaring GridSearchCV model

model = model_selection.GridSearchCV(

    estimator = clf,

    param_grid = param_grid,

    scoring = 'accuracy',

    verbose = 10,

    n_jobs = 1,

    cv = 5

)

#fitting values to the gridsearchcv model

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

####RandomizedSearchCV

```python
#defining a set of values as a dictionary for hyperparameters

param_grid = {

    "n_estimators":[100,200,300,400],

    "max_depth":[1,3,5,7],

    "reg_lambda":[.01,.1,.5]

}

#declaring RandomizedSearchCV model

model = model_selection.RandomizedSearchCV(

    estimator = clf,

    param_distributions = param_grid,

    scoring = 'accuracy',

    verbose = 10,

    n_jobs = 1,

    cv = 5,

    n_iter=10

)

#fitting values to the RandomizedSearchCV model

model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
